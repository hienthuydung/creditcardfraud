## Scale and undersample data for correlation matrix

```python
from sklearn.preprocessing import StandardScaler, RobustScaler

# RobustScaler is less prone to outliers.

std_scaler = StandardScaler()
rob_scaler = RobustScaler()

data['scaled_amount'] = rob_scaler.fit_transform(data['Amount'].values.reshape(-1,1))
data['scaled_time'] = rob_scaler.fit_transform(data['Time'].values.reshape(-1,1))

data.drop(['Time','Amount'], axis=1, inplace=True)



scaled_amount = data['scaled_amount']
scaled_time = data['scaled_time']

data.drop(['scaled_amount', 'scaled_time'], axis=1, inplace=True)
data.insert(0, 'scaled_amount', scaled_amount)
data.insert(1, 'scaled_time', scaled_time)

# Amount and Time are Scaled!

data.head()
```

```python
# Shuffle the data and select equal number of cases with class 0 and 1 respectively
data_shuffled = data.sample(frac=1)
data_reduced_fraud = data_shuffled[data_shuffled["Class"]==1]
data_reduced_nonfraud = data_shuffled[data_shuffled["Class"]==0][:len(data_reduced_fraud)]
data_reduced = pd.concat([data_reduced_fraud, data_reduced_nonfraud])
print ("Total sample size: {}".format(len(data_shuffled)))
print ("Fraud case reduced sample size: {}".format(len(data_reduced_fraud)))
print ("Non-fraud case reduced sample size: {}".format(len(data_reduced_nonfraud)))
print ("Reduced sample size: {}".format(len(data_reduced)), "\n")
```

```python
# plot correlation between features

corrmat = data.corr()

f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corrmat, square = True, cmap = 'coolwarm_r')
plt.title('Heatmap of Correlation (original dataset)', fontsize=15)
```

```python
# plot correlation between features of reduced dataset

corrmat = data_reduced.corr()

f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corrmat, square = True, cmap = 'coolwarm_r')
plt.title('Heatmap of Correlation (undersampled dataset)', fontsize=15)
```
